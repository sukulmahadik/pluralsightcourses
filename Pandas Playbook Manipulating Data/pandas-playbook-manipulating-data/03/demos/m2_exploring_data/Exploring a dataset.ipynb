```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# We read a dataset with weather data from The Netherlands
df = pd.read_csv('weather.csv')
```

```python
# Ask for the shape of our dataset
# This returns the number of rows and columns
# Sukul : Note that this is a property. Not a method.
df.shape
# Sukul : Below result shows that there are 8784 rows and 5 columns in the dataframe.
```

```python
# Ask for a more elaborate description of our data
# Including datatypes for each column, index type, memory usage
df.info()
# Sukul: Below tells that df is of the type DataFrame,the Number and type of columns, 
# no of non null values in each column, Number of columns of each datatype and memory usage.
# Also make a note that float64 type field cannot have a null value(missing value)
```

```python
# head() gets the first 5 rows (Default is 5 rows. However we can change the no of rows to see)
df.head()
# Or you can specify how many lines you want:
# df.head(50)
```

```python
# tail() gets the last 5 rows
df.tail()
```

```python
# Describe gives a statistical overview of our data (count,mean,std,min,max)
df.describe()
# Sukul: Non numeric columns may be ignore all together.
```

```python
# mean() calculates the mean per column. Note that such methods can work along with Series as well.
# When used with Series, it only gives the mean value for this series itself. 
# When used with dataframes, it gives the mean value for all eligible columns in the dataframe.
# Other similar methods are min,max,sum
df.mean()
```

```python
df.count()
```

```python
# max() calculates the maximum per column
df.max()
```

```python
# min() calculates the minimum per column
df.min()
```

```python
# median() calculates the median per column
df.median()
```

```python
df.std()
```

```python
df.var()
```

```python
df.skew()
```

```python
df.kurtosis()
```

```python
df.quantile()
```

```python
# These types of functions can also be called on a single column
# Here we calculate the minimum value in the pressure column

#Above we saw that when we invoked the function like min,max,sum on dataframe , the operation
# was performed for each of the eligible column. However when these can also be invoked on Series.
df['PRESSURE'].min()
```

```python
# mode() returns the most occurring value in the column
df['TEMP'].mode()
```

```python
#Value counts can only be executed on a Series.
df['TEMP'].value_counts()
```

```python
df.plot()
```

```python
# plotting a column is very easy
# note that we are invoking plot function directly on the Series.
df['TEMP'].plot()
```

```python
df['TEMP'].plot.hist()
```

```python
# There are several types of plot we can make
# This example shows a histogram for the temperature. 
# With histograms we have options of specifying bins to make more granular.
df['TEMP'].plot.hist(bins=100)
```
